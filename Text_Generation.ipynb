**DOWNLOADING DATASET**

```
from keras.utils.data_utils import get_file
path = get_file(
    'nietzsche.txt',
    origin='https://s3.amazonaws.com/text-datasets/nietzsche.txt')
```

```
import io
with io.open(path, encoding='utf-8') as f:
    text = f.read().lower()
print('corpus length:', len(text))
```

```
s = ""
for i in range(10, len(text)):
  if text[i] == '\n':
    s += (' ')
  else:
    s += text[i]
```

```
text = s
text[:100]
```

**PREPROCESSING DATA**

```
text.split(' ')[:100]
```

```
print("Number of words:", len(text.split(' ')))
```

```
unique_words = set(text.split(' '))
print('Number of unique words: ', len(unique_words))
unique_words = list(sorted(unique_words))
```

```
unique_words[0:10]
```

```
word_to_int = {}
int_to_word = {}
for index, word in enumerate(unique_words):
  word_to_int[word] = index
  int_to_word[index] = word
```

```
word_to_int.items()
```

```
int_to_word.items()
```

```
l = text.split(' ')
l1 = [word_to_int[i] for i in l]
len(l1)
```

```
import numpy as np
Tx = 70
X = []
for i in range(0, len(l1)-Tx, Tx):
  temp = []
  for j in range(i, i+Tx):
    temp.append(l1[j])
  X.append(temp)
X = np.array(X)
```

```
X.shape
```

```
m = X[0]
Tx = 70
```

```
import numpy as np
Tx = 70
Y = []
for i in range(1, len(l1)-Tx, Tx):
  temp = []
  for j in range(i, i+Tx):
    temp.append(l1[j])
  Y.append(temp)
Y = np.array(Y)
```

```
Y.shape
```

```
X[0]
```

```
Y[0]
```

**LABEL TO  ONE-HOT**

```
from keras.utils import to_categorical

Y = to_categorical(Y[:,:], num_classes = len(unique_words))
Y.shape
```

**BUILDING MODEL**

```
import keras
from keras.layers import Dense, Dropout, LSTM, Activation, Input
from keras.models import Model
from keras.layers.embeddings import Embedding
```

```
emb_matrix = np.zeros((len(unique_words)+1, 100))
emb_matrix.shape
```

```
embedding_layer = Embedding(len(unique_words)+1, 500, trainable = True, embeddings_initializer='uniform')
embedding_layer.build((None,))
```

```
def LSTM_model(input_shape):
    sentence_indices = Input(shape = input_shape, dtype = np.int32)
    embeddings = embedding_layer(sentence_indices)
    X = LSTM(256, return_sequences = True)(embeddings)
    X = Dropout(0.5)(X)
    X = LSTM(256, return_sequences = True)(X)
    X = Dropout(0.5)(X)
    X = LSTM(128, return_sequences = True)(X)
    X = Dropout(0.25)(X)
    X = Dense(len(unique_words), activation = 'softmax')(X)
    
    model = Model(inputs = sentence_indices, outputs = X)
    
    return model
```

```
model = LSTM_model((Tx,))
```

```
model.summary()
```

```
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```
model.fit(X, Y, epochs = 100, batch_size = 32, shuffle = True)
```
